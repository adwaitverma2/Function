#5. What are iterators in Python and how do they differ from iterables?
Iterable is an object that can be looped over (like list, tuple, set).

Iterator is an object that produces the next item in a sequence when next() is called.

iter() function is used to convert an iterable into an iterator.

Iterators maintain state and raise StopIteration when no items are left.

Example

nums = [1, 2, 3]      # iterable
it = iter(nums)       # iterator created

print(next(it))       # Output: 1
print(next(it))       # Output: 2

#6. Explain the concept of generators in Python and how they are defined.
Generators are functions that return an iterator and allow iteration over a sequence of values lazily, one at a time.

They are defined like normal functions but use the yield keyword instead of return.

Each time the generator function is called using next(), it resumes from where it left off.

Generators are memory efficient, especially for large data sets.

Example

def countdown(n):
    while n > 0:
        yield n
        n -= 1

for i in countdown(3):
    print(i)

Output: 3, 2, 1

#7. What are the advantages of using generators over regular functions?
Memory Efficient: Generators don’t store all values in memory; they yield one at a time.

Lazy Evaluation: Values are produced only when requested, not all at once.

Improved Performance: Suitable for large datasets or streams.

Simplifies Code: Cleaner syntax for iterators using yield.

Example

# Generator version
def gen_numbers():
    for i in range(1000000):
        yield i

# List version
def list_numbers():
    return [i for i in range(1000000)]

#10. What is the difference between map(), reduce(), and filter() functions in Python?

map -> Transforms each element using a function.

Example: map(lambda x: x*2, [1,2,3]) → [2,4,6]

filter -> Selects elements for which a function returns True.

Example: filter(lambda x: x%2==0, [1,2,3]) → [2]

reduce -> Repeatedly applies a function to elements, reducing them to a single result.

Requires from functools import reduce.

Example: reduce(lambda x, y: x+y, [1,2,3]) → 6

map() and filter() return iterables; reduce() returns a single value.

Example

from functools import reduce

nums = [1, 2, 3, 4]

mapped = list(map(lambda x: x * 2, nums))

filtered = list(filter(lambda x: x % 2 == 0, nums))

reduced = reduce(lambda x, y: x + y, nums)

print(mapped)   # [2, 4, 6, 8]

print(filtered) # [2, 4]

print(reduced)  # 10

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
a = [1, 2, 3, 4, 5, 6]
def sumOfEvenNumber(a):
    sum = 0
    for i in a:
        if i % 2 == 0:
            sum = sum + i
    return sum
print(sumOfEvenNumber(a))
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, x):
        self.x = x  # total number of terms to generate
        self.count = 0              # to track how many terms have been returned
        self.a = 0                  # first Fibonacci number
        self.b = 1                  # second Fibonacci number

    def __iter__(self):
        return self  # return the iterator object

    def __next__(self):
        if self.count >= self.x:
            raise StopIteration  # stop when limit is reached

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            z = self.a + self.b  # next term is sum of previous two
            self.a, self.b = self.b, z  # update values
            self.count += 1
            return z
n = int(input("Enter number"))
fib = Fibonacci(n)

print("Fibonacci seq")
for num in fib:
    print(num)




```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powerOfTwoExp(max_exp):
    for i in range(max_exp + 1):  # loop from 0 to max_exp
        yield 2 ** i
for num in powerOfTwoExp(5):
    print(num)
```
